# **코딩변경**

기존의 데이터의 값을 새로운 기준으로 만드는 과정

```python
import pandas as pd
```

```python
df=pd.read_csv('cosmetics_.csv', encoding='utf-8')
```

```python
df.head()
```

```python
df['edu'].value_counts(sort=False) # 값_빈도   # () 가로로 묶어줌 -> 집단별로 # 빈도수 sort=false 순서대로
```

```python
def edu_groups(series):# edu_group 인위적인 이름(연속적인 값) <- 분류위한 임시적인 객체
  if series==2:      # = 할당  == 같다
    return 1
  elif series==3 or series==4:
    return 2
  elif series >=5:
    return 3
```

```python
df['eduM']=df['edu'].apply(edu_groups)   
# apply 붙인다  edu 라는 변수를 가지고 위에서 선언한 변수 찾아서 eduM할당
```

```python
df['eduM'].value_counts(sort=False)
```
