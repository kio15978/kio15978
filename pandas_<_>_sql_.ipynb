# **join**

```python
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],'value': np.random.randn(4)})
df2 = pd.DataFrame({'key': ['B', 'D', 'D', 'E'],'value': np.random.randn(4)})
```

# **11. df1과 df2를 inner join, df1 키와 df2 키를 기준**

select *

from df1 inner join df2

on df1.key=df2.key;

-

select *

from df1, df2

where df1.key=df2.key;

```python
pd.merge(df1,df2,on='key')
```

# **12. df1과 df2를 left join, df1 키와 df2 키를 기준**

select *

from df1 left join df2

on df1.key=df2.key;

```python
pd.merge(df1,df2,on='key',how='left')
```

# **13. df1과 df2를 right join, df1 키와 df2 키를 기준**

select *

from df1 right join df2

on df1.key=df2.key;

```python
pd.merge(df1,df2,on='key',how='right')
```

# **union**

```python
df1 = pd.DataFrame({'city': ['Chicago', 'SanFrancisco', 'New York City'],'rank': range(1, 4)})
df2 = pd.DataFrame({'city': ['Chicago', 'Boston', 'LosAngeles'],'rank': [1, 4, 5]})
```

# **14. df1과 df2를 붙여라**

select city, rank

from df1


union

select city, rank

from df2;

```python
pd.concat([df1,df2])
```

# **15. df1과 df2를 중복된 값 없이 붙여라**

select city, rank

from df1

union all

select city, rank

from df2;

```python
pd.concat([df1,df2]).drop_duplicates()
```
